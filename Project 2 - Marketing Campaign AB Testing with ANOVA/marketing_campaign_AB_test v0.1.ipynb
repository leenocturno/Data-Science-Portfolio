```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('WA_Marketing-Campaign.csv')
```

Construct a function to retrieve descriptive information at at once.

```python
def get_info(df):
    print('----------------------------- First 5 Rows -----------------------------')
    display(pd.DataFrame(df.head()))
    print('----------------------------- 5 Random Rows -----------------------------')
    display(pd.DataFrame(df.sample(5)))
    print('----------------------------- Shape -----------------------------')
    print(df.shape)
    print('----------------------------- Value Counts -----------------------------')
    for index, col in enumerate(df.columns):
        if not df[col].dtypes in (['int64','float64']):
            vc = pd.DataFrame(df[col].value_counts()).reset_index()
    display(vc)
    print('----------------------------- Description -----------------------------')
    display(df.describe([0, 0.05, 0.50, 0.95, 0.99, 1]).iloc[:,2:].T)
    print('----------------------------- Info -----------------------------')
    print(df.info())
```

```python
get_info(df)
```

```python
#further examine the columns
print(df['Promotion'].value_counts())
print(df['week'].value_counts())
print(df['AgeOfStore'].min())
print(df['AgeOfStore'].max())
```

<b><u>Columns</u></b><br>

**MarketID:** the unique identifier of the campaign

**MarketSize:** size of the market by sales. Has 3 unique values: large, medium, and small

**LocationID:** the unique identifier of the store location

**AgeOfStore:** the age of store in years. Max age is 28 years and min age is 1 year old.

**Promotion:** Contains 3 unique values: 1, 2, and 3

**week:** one of four weeks when the promotions were run

**SalesInThousands:** sales amount for a specific LocationID, Promotion, and week


```python
df.groupby(['MarketSize','Promotion']).agg({'SalesInThousands':['mean','sum']})
```
